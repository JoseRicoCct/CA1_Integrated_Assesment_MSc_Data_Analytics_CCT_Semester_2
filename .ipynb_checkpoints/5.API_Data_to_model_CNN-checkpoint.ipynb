```python
import keras
import time
from keras import layers
from keras.datasets import imdb
from keras.preprocessing import sequence
max_features = 2000
max_len = 500
```

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)
x_train = sequence.pad_sequences(x_train, maxlen=max_len)
x_test = sequence.pad_sequences(x_test, maxlen=max_len)
```

```python
#  CNN structure
model = keras.models.Sequential()
model.add(layers.Embedding(max_features, 128,
input_length=max_len,
name='embed'))
model.add(layers.Conv1D(32, 7, activation='relu'))
model.add(layers.MaxPooling1D(5))
model.add(layers.Conv1D(32, 7, activation='relu'))
model.add(layers.GlobalMaxPooling1D())
model.add(layers.Dense(1))
model.summary()
```

```python
# Compile the model
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Train the model
start_time = time.time()
history = model.fit(x_train, y_train,epochs=20,batch_size=128,validation_split=0.2)
end_time = time.time()
```

```python
execution_time = end_time - start_time

# Convert execution time to minutes and seconds
minutes = int(execution_time // 60)
seconds = int(execution_time % 60)

print(f"Total execution time: {minutes} minutes {seconds} seconds")
```

```python
import keras
from keras import layers, regularizers
from keras.datasets import imdb
from keras.preprocessing import sequence

# Configuration options:
max_features = 2000
max_len = 500
embedding_size = 64
kernel_size = 3
pool_size = 5
filters = 16
dense_units = 16
dropout_rate = 0.4
l2_penalty = 0.001

# Load and preprocess data:
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)
x_train = sequence.pad_sequences(x_train, maxlen=max_len)
x_test = sequence.pad_sequences(x_test, maxlen=max_len)

# Define the model:
model = keras.models.Sequential()

# Add an Embedding layer with dropout:
model.add(layers.Embedding(max_features, embedding_size, input_length=max_len))
model.add(layers.Dropout(dropout_rate))

# Add a Conv1D layer with L2 regularization:
model.add(layers.Conv1D(filters, kernel_size, activation='relu', 
                        kernel_regularizer=regularizers.l2(l2_penalty)))

# Add a MaxPooling1D layer:
model.add(layers.MaxPooling1D(pool_size))

# Add a GlobalMaxPooling1D layer to reduce the number of features:
model.add(layers.GlobalMaxPooling1D())

# Add a Dense layer with L2 regularization and dropout:
model.add(layers.Dense(dense_units, activation='relu', 
                       kernel_regularizer=regularizers.l2(l2_penalty)))
model.add(layers.Dropout(dropout_rate))

# Add the output layer with sigmoid activation for binary classification:
model.add(layers.Dense(1, activation='sigmoid'))

# Model summary:
model.summary()
```

```python
# Compile the model:
model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Train the model
start_time = time.time()
history = model.fit(x_train, y_train,epochs=20,batch_size=64, validation_split=0.2)
end_time = time.time()
```

```python
execution_time = end_time - start_time

# Convert execution time to minutes and seconds
minutes = int(execution_time // 60)
seconds = int(execution_time % 60)

print(f"Total execution time: {minutes} minutes {seconds} seconds")
```
