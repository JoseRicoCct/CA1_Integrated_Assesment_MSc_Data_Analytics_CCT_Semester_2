```python
import pandas as pd

def count_genders(df, gender_column='Sex'):
    """
    Count the number of males and females in a pandas DataFrame.

    Parameters:
    df (pandas.DataFrame): The DataFrame containing the gender data.
    gender_column (str): The name of the column containing gender information.

    Returns:
    dict: A dictionary with counts for 'Male' and 'Female'.
    """
    gender_counts = df[gender_column].value_counts()
    return gender_counts.to_dict()

# Assuming 'df' is your DataFrame
# Here's how you would use the function:
gender_counts = count_genders(df)
print(gender_counts)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Encoding "Sex" as binary values
encoder = LabelEncoder()
df['Sex'] = encoder.fit_transform(df['Sex'])  # Male will be 1, Female will be 0

# One-hot encoding "Job Title"
onehot_encoder = OneHotEncoder(sparse=False)
job_title_encoded = onehot_encoder.fit_transform(df['Job Title'].values.reshape(-1, 1))

# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(job_title_encoded, df['Sex'], test_size=0.2, random_state=42)

# Building the model
model = Sequential([
    Dense(64, input_dim=X_train.shape[1], activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Training the model
model.fit(X_train, y_train, epochs=100, batch_size=5, verbose=1)

# Evaluating the model
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f'Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}')
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Load your data into 'df'

# Encode 'Sex' as a binary variable
sex_encoder = LabelEncoder()
df['Sex'] = sex_encoder.fit_transform(df['Sex'])  # Male will be 1, Female will be 0

# Encode 'Job Title' with integers
job_title_encoder = LabelEncoder()
df['Job Title Encoded'] = job_title_encoder.fit_transform(df['Job Title'])

# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(df[['Job Title Encoded']], df['Sex'], test_size=0.2, random_state=42)

# Parameters
input_dim = df['Job Title Encoded'].max() + 1  # Add 1 because index starts from 0
embedding_dim = 50  # You can tune this

# Neural network with an embedding layer
job_title_input = Input(shape=(1,), name='job_title_input')
job_title_embedding = Embedding(input_dim=input_dim, output_dim=embedding_dim, name='job_title_embedding')(job_title_input)
job_title_vec = Flatten(name='flatten_job_titles')(job_title_embedding)

# Hidden layers
hidden = Dense(64, activation='relu')(job_title_vec)
hidden = Dropout(0.4)(hidden)  # Add dropout for regularization
hidden = Dense(32, activation='relu')(hidden)
hidden = Dropout(0.4)(hidden)

# Output layer
output = Dense(1, activation='sigmoid')(hidden)

# Create and compile model
model = Model(inputs=job_title_input, outputs=output)
model.compile(optimizer=Adam(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])

# Train
model.fit(X_train, y_train, epochs=10, batch_size=256, validation_data=(X_test, y_test), verbose=1)  # Fewer epochs for quicker convergence

# Evaluate
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f'Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}')
```

```python
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from scipy.sparse import hstack

# Assuming 'df' has been loaded and preprocessed

# Ensure 'Sex' is encoded as an integer
df['Sex'] = LabelEncoder().fit_transform(df['Sex'])

# One-hot encoding 'Job Title' with sparse output and converting 'Age' to sparse
onehot_encoder = OneHotEncoder(sparse=True)
job_title_encoded_sparse = onehot_encoder.fit_transform(df['Job Title'].values.reshape(-1, 1))
age_sparse = scipy.sparse.csr_matrix(df[['Age']])
X_sparse = hstack([job_title_encoded_sparse, age_sparse])

# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X_sparse, df['Sex'], test_size=0.2, random_state=42)

# Generator function to yield batches from sparse matrix
def sparse_to_dense(X_sparse, y, batch_size=100):
    num_samples = X_sparse.shape[0]
    for start_idx in range(0, num_samples, batch_size):
        end_idx = start_idx + batch_size
        batch_x = X_sparse[start_idx:end_idx].toarray()  # Convert sparse batch to dense
        batch_y = y[start_idx:end_idx]
        yield batch_x, batch_y

# Create generators for training and testing
train_gen = sparse_to_dense(X_train, y_train, batch_size=100)
test_gen = sparse_to_dense(X_test, y_test, batch_size=100)

# Define the model
input_dim = X_train.shape[1]
model = Sequential([
    Dense(64, input_dim=input_dim, activation='relu'),
    Dropout(0.5),
    Dense(32, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

# Calculate steps per epoch and validation steps
train_steps = np.ceil(X_train.shape[0] / 100).astype(int)
test_steps = np.ceil(X_test.shape[0] / 100).astype(int)

# Debugging: Inspect the first batch
first_batch_x, first_batch_y = next(train_gen)
print("First batch X shape:", first_batch_x.shape)
print("First batch Y shape:", first_batch_y.shape)
print("First batch Y type:", first_batch_y.dtype)

# Train the model
model.fit(train_gen, steps_per_epoch=train_steps, epochs=10, verbose=1, validation_data=test_gen, validation_steps=test_steps)

# Evaluate the model
loss, accuracy = model.evaluate(test_gen, steps=test_steps)
print(f'Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}')
```
