```python
import time
from IPython.utils import io
import pandas as pd
start_time = time.time()

# Replace 'file_path' with the path to your CSV file
file_path = "C:/Users/jose/Downloads/people.csv"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
df.head()
```

```python
df.info()
```

```python
unique_job_titles = df['Job Title'].unique()
print("Distinct count of job titles:", len(unique_job_titles))
```

```python
# Group by 'Job Title' and 'Sex', and aggregate counts
job_title_counts = df.groupby(['Job Title', 'Sex']).size().reset_index(name='Count')

# Calculate total count for each job title
job_title_total_counts = df.groupby('Job Title').size().reset_index(name='Total Count')

# Merge total counts with the counts by sex
job_title_counts = pd.merge(job_title_counts, job_title_total_counts, on='Job Title')

# Pivot the DataFrame to have "Male" and "Female" as columns
job_title_pivot = job_title_counts.pivot_table(index='Job Title', columns='Sex', values='Count', fill_value=0)

# Add a column for Total Count
job_title_pivot['Total Count'] = job_title_pivot.sum(axis=1)

# Reset the index to make 'Job Title' a column again
job_title_pivot.reset_index(inplace=True)



# Rearrange the columns
job_title_pivot = job_title_pivot[['Job Title', 'Male', 'Female', 'Total Count']]

# Sort the DataFrame by "Total Count" in descending order
job_title_pivot = job_title_pivot.sort_values(by='Total Count', ascending=False).reset_index(drop=True).rename_axis(None, axis=1)

# Display the pivoted and sorted DataFrame

job_title_pivot
```

```python
def copy_until_gb(df):
    target_size = 6 * df.memory_usage(deep=True).sum()  # Target size is five times the original DataFrame's size
    copied_df = pd.DataFrame(columns=df.columns)

    while copied_df.memory_usage(deep=True).sum() <= target_size:
        copied_df = pd.concat([copied_df, df], ignore_index=True)

    return copied_df

# Example usage assuming your DataFrame is named 'df'
copied_df = copy_until_gb(df)
```
